```python
rooms = {
    101: None,
    102: None,
    103: None,
    104: None,
    105: None,
}

occupied = set()

while True:
    print("\n1. Allot Room")
    print("2. Vacate Room")
    print("3. Room Info")
    print("4. Exit")

    choice = input("Enter choice: ")
    if choice == '1':
        room_no = int(input("Enter room number to allot: "))

        if room_no in rooms:
            if rooms[room_no] is None:
                name = input("Enter customer name: ")
                rooms[room_no] = name
                occupied.add(room_no)
                print(f"Room {room_no} allotted to {name}")
            else:
                print("Room already occupied!")
        else:
            print("Invalid room number!")


    elif choice == '2':
        room_no = int(input("Enter room number to vacate: "))

        if room_no in rooms:
            if rooms[room_no] is not None:
                print(f"Room {room_no} vacated (Customer: {rooms[room_no]})")
                rooms[room_no] = None
                occupied.remove(room_no)
            else:
                print("Room already empty!")
        else:
            print("Invalid room number!")


    elif choice == '3':
        for room_no, name in rooms.items():
            if name is None:
                print(f"Room {room_no} → Empty")
            else:
                print(f"Room {room_no} → Occupied by {name}")

        print("Occupied Rooms Set:", occupied)


    elif choice == '4':
        print("Exiting system...")
        break

    else:
        print("Invalid choice!")
```

```python
accounts = {}

def create_account():
    userid = input("Enter new User ID: ")
    if userid in accounts:
        print("Account already exists")
        return
    password = input("Enter new Password: ")
    accounts[userid] = {"password": password, "balance": 0}
    print("Account created")

def login():
    userid = input("Enter User ID: ")
    password = input("Enter Password: ")
    if userid in accounts and accounts[userid]["password"] == password:
        print("Login successful")
        return userid
    else:
        print("Invalid credentials")
        return None

def check_balance(userid):
    print("Balance:", accounts[userid]["balance"])

def deposit(userid):
    amt = int(input("Enter amount to deposit: "))
    accounts[userid]["balance"] += amt
    print("Amount deposited")

def withdraw(userid):
    amt = int(input("Enter amount to withdraw: "))
    if accounts[userid]["balance"] >= amt:
        accounts[userid]["balance"] -= amt
        print("Withdrawal successful")
    else:
        print("Insufficient balance")

while True:
    print("\n1. Create Account")
    print("2. Login")
    print("3. Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        create_account()

    elif choice == "2":
        user = login()
        if user:
            while True:
                print("\n1. Check Balance")
                print("2. Deposit")
                print("3. Withdraw")
                print("4. Logout")
                option = input("Enter choice: ")

                if option == "1":
                    check_balance(user)
                elif option == "2":
                    deposit(user)
                elif option == "3":
                    withdraw(user)
                elif option == "4":
                    break
                else:
                    print("Invalid option")

    elif choice == "3":
        print("Thank you")
        break

    else:
        print("Invalid choice")
```

```python
import time

parking = {}
total_slots = 5

def entry(vehicle):
    if len(parking) < total_slots:
        parking[vehicle] = time.time()
        print("Vehicle parked")
    else:
        print("Parking Full")

def exit(vehicle):
    if vehicle in parking:
        entry_time = parking[vehicle]
        hours = (time.time() - entry_time) / 3600
        fee = int(hours + 1) * 10
        print("Parking Fee:", fee)
        del parking[vehicle]
    else:
        print("Vehicle not found")

def available_slots():
    print("Available slots:", total_slots - len(parking))

entry("OD02AB1234")
available_slots()
exit("OD02AB1234")
available_slots()
```

```python
tasks = []

def add_task():
    task = input("Enter task: ")
    deadline = input("Enter deadline (YYYY-MM-DD): ")
    priority = input("Enter priority (high/medium/low): ").lower()
    category = input("Enter category: ")
    tasks.append({'task': task, 'deadline': deadline, 'priority': priority, 'category': category})
    print("Task added")

def view_tasks():
    if not tasks:
        print("No tasks")
        return
    # sort by priority, high first
    priority_order = {'high': 3, 'medium': 2, 'low': 1}
    sorted_tasks = sorted(tasks, key=lambda x: priority_order.get(x['priority'], 0), reverse=True)
    for t in sorted_tasks:
        print(f"Task: {t['task']}, Deadline: {t['deadline']}, Priority: {t['priority']}, Category: {t['category']}")

def filter_by_category():
    cat = input("Enter category to filter: ")
    filtered = [t for t in tasks if t['category'].lower() == cat.lower()]
    if not filtered:
        print("No tasks in this category")
        return
    for t in filtered:
        print(f"Task: {t['task']}, Deadline: {t['deadline']}, Priority: {t['priority']}")

while True:
    print("\n1. Add Task")
    print("2. View All Tasks")
    print("3. Filter by Category")
    print("4. Exit")
    choice = input("Enter choice: ")
    if choice == '1':
        add_task()
    elif choice == '2':
        view_tasks()
    elif choice == '3':
        filter_by_category()
    elif choice == '4':
        break
    else:
        print("Invalid choice")
```

```python
grades = {}

def add_grade():
    try:
        name = input("Enter name: ")
        if name == "":
            raise ValueError("Name cannot be empty")

        grade = float(input("Enter grade: "))
        grades[name] = grade

        print("Grade added successfully")

    except ValueError as e:
        print("Error:", e)


def delete_grade():
    try:
        name = input("Enter name to delete: ")

        if name in grades:
            del grades[name]
            print("Grade deleted successfully")
        else:
            raise KeyError("Name not found")

    except KeyError as e:
        print("Error:", e)


add_grade()

print(grades)
```

```python
contacts = {}

def add_contact():
    try:
        name = input("Enter name: ")
        phone = input("Enter phone number: ")

        if name == "" or phone == "":
            raise ValueError("Fields cannot be empty")

        if name in contacts:
            raise ValueError("Contact already exists")

        if not phone.isdigit() or len(phone) != 10:
            raise ValueError("Invalid phone number")

        contacts[name] = phone
        print("Contact added")

    except ValueError as e:
        print("Error:", e)


def edit_contact():
    try:
        name = input("Enter name to edit: ")

        if name not in contacts:
            raise KeyError("Contact not found")

        phone = input("Enter new phone number: ")

        if not phone.isdigit() or len(phone) != 10:
            raise ValueError("Invalid phone number")

        contacts[name] = phone
        print("Contact updated")

    except (KeyError, ValueError) as e:
        print("Error:", e)


def search_contact():
    try:
        name = input("Enter name to search: ")

        if name == "":
            raise ValueError("Name cannot be empty")

        if name not in contacts:
            raise KeyError("Contact not found")

        print(name, ":", contacts[name])

    except (KeyError, ValueError) as e:
        print("Error:", e)


while True:
    print("\n1.Add Contact")
    print("2.Edit Contact")
    print("3.Search Contact")
    print("4.Exit")

    ch = input("Enter choice: ")

    if ch == "1":
        add_contact()
    elif ch == "2":
        edit_contact()
    elif ch == "3":
        search_contact()
    elif ch == "4":
        break
    else:
        print("Invalid choice")
```

```python
import time
import random

accounts = {
    "1001": 5000,
    "1002": 3000,
    "1003": 7000
}

class TransactionTimeout(Exception):
    pass

def transfer_money():
    try:
        sender = input("Enter sender account number: ")
        receiver = input("Enter receiver account number: ")

        if sender not in accounts or receiver not in accounts:
            raise KeyError("Invalid account number")

        amount = float(input("Enter amount to transfer: "))

        if accounts[sender] < amount:
            raise ValueError("Insufficient balance (Overdraft)")

        print("Processing transaction...")
        time.sleep(2)

        if random.choice([True, False]):
            raise TransactionTimeout("Transaction timed out. Please try again.")

        accounts[sender] -= amount
        accounts[receiver] += amount

        print("Transaction successful!")

    except KeyError as e:
        print("Error:", e)

    except ValueError as e:
        print("Error:", e)

    except TransactionTimeout as e:
        print("Error:", e)

    except Exception as e:
        print("Unexpected error:", e)

def show_accounts():
    print("\nAccount Balances")
    for acc, bal in accounts.items():
        print(acc, ":", bal)

while True:
    print("\n--- Banking System ---")
    print("1. Transfer Money")
    print("2. Show Accounts")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        transfer_money()
    elif choice == "2":
        show_accounts()
    elif choice == "3":
        break
    else:
        print("Invalid choice")
```
