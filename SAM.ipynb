```python
#allot room
rm={
101:None,
102:None,
103:None,
104:None,
105:None
} 
ocuupied = set()
x=int(input("Enter number of members to participate"))
for i in range(x):
    na=input("Enter the name of member")
    name=na.lower()
    if name not in ocuupied:
        for room, occupant in rm.items():
            if occupant is None:
                rm[room] = name
                ocuupied.add(name)
                print(f"{name} has been allotted room {room}")
                break
    else:
        print(f"{name} is already allotted a room.")
        
```

```python
rooms = {
    101: None,
    102: None,
    103: None,
    104: None,
    105: None,
}

occupied = set()

while True:
    print("\n1. Allot Room")
    print("2. Vacate Room")
    print("3. Room Info")
    print("4. Exit")

    choice = input("Enter choice: ")
    if choice == '1':
        room_no = int(input("Enter room number to allot: "))

        if room_no in rooms:
            if rooms[room_no] is None:
                name = input("Enter customer name: ")
                rooms[room_no] = name
                occupied.add(room_no)
                print(f"Room {room_no} allotted to {name}")
            else:
                print("Room already occupied!")
        else:
            print("Invalid room number!")


    elif choice == '2':
        room_no = int(input("Enter room number to vacate: "))

        if room_no in rooms:
            if rooms[room_no] is not None:
                print(f"Room {room_no} vacated (Customer: {rooms[room_no]})")
                rooms[room_no] = None
                occupied.remove(room_no)
            else:
                print("Room already empty!")
        else:
            print("Invalid room number!")


    elif choice == '3':
        for room_no, name in rooms.items():
            if name is None:
                print(f"Room {room_no} → Empty")
            else:
                print(f"Room {room_no} → Occupied by {name}")

        print("Occupied Rooms Set:", occupied)


    elif choice == '4':
        print("Exiting system...")
        break

    else:
        print("Invalid choice!")
```

```python
uid = "101"
pwd = "123"
balance = 5000

def login():
    userid = input("Enter your id: ")
    password = input("Enter your password: ")
    
    if pwd == password and userid == uid:
        print("Login suc")
        return True
    else:
        print("Invalid ")
        return False


def bal():
    global balance
    print("Your available balance is:", balance)


def deposit():
    global balance
    amt = int(input("Enter deposit amount: "))
    balance = balance + amt
    print("Amount deposited succ")


def withdraw():
    global balance
    amt = int(input("Enter your withdrawal amt: "))
    
    if balance >= amt:
        balance = balance - amt
        print("Withdrawal succ")
    else:
        print("Insufficient bal")


def thank():
    print("Thank you ")


def choice():
    print("\n1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")


if login():
    while True:
        choice()
        option = input("Enter your choice: ")
        
        if option == "1":
            bal()
        elif option == "2":
            deposit()
        elif option == "3":
            withdraw()
        elif option == "4":
            thank()
            break
        else:
            print("Invalid option")
```

```python
accounts = {}

def create_account():
    userid = input("Enter new User ID: ")
    if userid in accounts:
        print("Account already exists")
        return
    password = input("Enter new Password: ")
    accounts[userid] = {"password": password, "balance": 0}
    print("Account created")

def login():
    userid = input("Enter User ID: ")
    password = input("Enter Password: ")
    if userid in accounts and accounts[userid]["password"] == password:
        print("Login successful")
        return userid
    else:
        print("Invalid credentials")
        return None

def check_balance(userid):
    print("Balance:", accounts[userid]["balance"])

def deposit(userid):
    amt = int(input("Enter amount to deposit: "))
    accounts[userid]["balance"] += amt
    print("Amount deposited")

def withdraw(userid):
    amt = int(input("Enter amount to withdraw: "))
    if accounts[userid]["balance"] >= amt:
        accounts[userid]["balance"] -= amt
        print("Withdrawal successful")
    else:
        print("Insufficient balance")

while True:
    print("\n1. Create Account")
    print("2. Login")
    print("3. Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        create_account()

    elif choice == "2":
        user = login()
        if user:
            while True:
                print("\n1. Check Balance")
                print("2. Deposit")
                print("3. Withdraw")
                print("4. Logout")
                option = input("Enter choice: ")

                if option == "1":
                    check_balance(user)
                elif option == "2":
                    deposit(user)
                elif option == "3":
                    withdraw(user)
                elif option == "4":
                    break
                else:
                    print("Invalid option")

    elif choice == "3":
        print("Thank you")
        break

    else:
        print("Invalid choice")
```

```python
grade_book = {}

n = int(input("Enter number of students: "))

for i in range(n):
    name = input("\nEnter student name: ")
    
    m1 = int(input("Enter marks for Subject 1: "))
    m2 = int(input("Enter marks for Subject 2: "))
    m3 = int(input("Enter marks for Subject 3: "))
    
    total = m1 + m2 + m3
    average = total / 3
    
    if average >= 90:
        grade = "A"
    elif average >= 75:
        grade = "B"
    elif average >= 60:
        grade = "C"
    elif average >= 40:
        grade = "D"
    else:
        grade = "Fail"
    
    grade_book[name] = {
        "Total": total,
        "Average": average,
        "Grade": grade
    }

print("\n------ Grade Book ------")

for name, details in grade_book.items():
    print("\nName:", name)
    print("Total:", details["Total"])
    print("Average:", details["Average"])
    print("Grade:", details["Grade"])
```
